Tokenization is the process of tokenizing or splitting a string, text into a list of tokens. One can think of token as parts like a word is a token in a sentence, and a sentence is a token in a paragraph.

<h3> White Space Tokenization </h3>

```python
def whitespace_tokenize(x):
  # import WhitespaceTokenizer() method from nltk
  from nltk.tokenize import WhitespaceTokenizer

  # Create a reference variable for Class WhitespaceTokenizer
  tk = WhitespaceTokenizer()
	
  # Use tokenize method
  y = tk.tokenize(x)
	
  print('='*26,'WHITE SPACE TOKENIZER','='*26,'\n')
  print('\t',y)

whitespace_tokenize("I am R.Sailaja")
```

<h3> Dictionary Based Tokenization </h3>

```python
from nltk.tokenize import regexp_tokenize

sent = "I like apple fruit but grape fruit more"
dict_list = ["apple fruit", "grape fruit"]
newdict = {}
for item in dict_list:
    dk = item.replace(" ", "_")
    newdict[item] = dk

for key, val in newdict.items():
    if key in sent:
        sent = sent.replace(key, val)

res = regexp_tokenize(sent, pattern='\S+')
print(res)
```

<H3> SUB WORD TOKENIZATION </H3>

Subword tokenization allows the model to have a reasonable vocabulary size while being able to learn meaningful context-independent representations.

```python
from nltk.tokenize import sent_tokenize,word_tokenize
txt="Tokenization in Python is the most primary step in any natural language processing program. This may find its utility in statistical analysis, parsing, spell-checking, counting and corpus generation etc. The tokenizer is a Python (2 and 3) module."
print('='*80,'WORD TOKENIZER','='*80,'\n')
print('\t',word_tokenize(txt))
print('='*80,'SENTANCE TOKENIZER','='*80,'\n')
print('\t',sent_tokenize(txt))
```

```python
# General Word based tokenization
text = """ Tesla, Inc. is an American automotive and energy company based in PAlo Alto, California. 
The company specializes in electric car manufacturing and, through its SolarCity subsidiary, solar panel manufacturing. """
data = text.split()
for i in data:
  print (i)
```

<center><h3> Using Moses Tokenizer </h3></center>

```python
!pip install mosestokenizer
from mosestokenizer import *
```

```python
# Word tokenize
tokenize = MosesTokenizer('en')
tokenize('Hello World!')
```

```python
with MosesDetokenizer('en') as detokenize:
  x=detokenize(['Hello', 'World', '!'])
x
```

```python
with MosesSentenceSplitter('en') as splitsents:
  x=splitsents(['Mr. Smith is away.  Do you want to','leave a message?'])
x
```

```python
with MosesPunctuationNormalizer('en') as normalize:
  x=normalize('«Hello World» — she said…')
x
```

```python
from nltk.tokenize import regexp_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
s = "I can't do this. I won't do it!"
print('='*40,'WORD TOKENIZER','='*40,'\n')
print('\t',sent_tokenize(s))
print('='*40,'SENTANCE TOKENIZER','='*40,'\n')
print'\t',(word_tokenize(s))
print('='*40, 'REG_EX TOKENIZER','='*40,'\n')
print('\t',regexp_tokenize(s,"[\w']+"))
```
